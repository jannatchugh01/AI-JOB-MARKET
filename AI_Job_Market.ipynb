# **KPI**

```python
# KPI Calculations

total_jobs = df.shape[0]
avg_salary = df['annual_salary_usd'].mean()
highest_salary = df['annual_salary_usd'].max()
lowest_salary = df['annual_salary_usd'].min()
total_countries = df['country'].nunique()
remote_jobs = (df['remote_work'] == 'Remote').sum()
remote_percent = (remote_jobs / total_jobs) * 100
llm_roles = df['is_llm_role'].sum()
avg_demand = df['demand_score'].mean()

print("="*70)
print("           AI JOB MARKET DASHBOARD (2025–2026)")
print("="*70)

print(f"📌 Total Job Postings      : {total_jobs}")
print(f"💰 Average Salary          : ${avg_salary:,.0f}")
print(f"🏆 Highest Salary          : ${highest_salary:,.0f}")
print(f"📉 Lowest Salary           : ${lowest_salary:,.0f}")
print(f"🌍 Countries Covered       : {total_countries}")
print(f"🏠 Remote Jobs             : {remote_percent:.1f}%")
print(f"🤖 LLM Roles               : {llm_roles}")
print(f"📈 Average Demand Score    : {avg_demand:.2f}")
```

## **ANALYSIS**

```python
# Job Title Distribution

plt.figure(figsize=(10,6))

ax = (df['job_title'].value_counts().plot(kind='barh'))
plt.title("Job Title Distribution")
plt.xlabel("Job Count")
plt.ylabel("Job Title")
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')

plt.tight_layout()
plt.show()
```

### **Insight:**
The job title distribution shows which titles dominate current AI hiring. Roles such as LLM Engineer and Robotics Engineer(AI) typically account for the largest share, reflecting their broad applicability. Niche titles (e.g., computer vision engineer, Ai research scientist) have fewer postings.

```python
# Salary Distribution

plt.figure(figsize=(10,6))

ax = (df.groupby('job_title')['annual_salary_usd'].mean().sort_values().plot(kind='barh'))
plt.title("Average Salary Distribution")
plt.xlabel("Avg Salary")
plt.ylabel("Job Title")
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')

plt.tight_layout()
plt.show()
```

### **Insight:**
Specialist roles (e.g., AI Solution Architect, Senior ML Engineer) command higher average compensation than generalist titles. Use this chart to benchmark a specific offer — roles below the overall average may have upside through equity or faster growth trajectories.

```python
# Salary Distribution among different factors

plt.figure(figsize=(16,10))

# Plot 1
plt.subplot(2,2,1)
sns.boxplot(x='experience_level', y='annual_salary_usd', data=df)
plt.title("Annual Salary Distribution by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Annual Salary (USD)")

# Plot 2
plt.subplot(2,2,2)
sns.boxplot(x='education_required', y='annual_salary_usd', data=df)
plt.title("Annual Salary Distribution by Education Level")
plt.ylabel("Annual Salary (USD)")
plt.xlabel("Education")

# Plot 3
plt.subplot(2,2,3)
ax = (df.groupby('country')['annual_salary_usd'].mean().sort_values().plot(kind='barh'))
plt.title("Average Salary by Country")
plt.xlabel("Average Salary (USD)")
plt.ylabel("Country")
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')

# Plot 4
plt.subplot(2,2,4)
ax = (df.groupby('company_size')['annual_salary_usd'].mean().sort_values().plot(kind='barh'))
plt.title("Average Salary by Company Size")
plt.xlabel("Average Salary (USD)")
plt.ylabel("Company Size")
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')

plt.tight_layout()
plt.show()
```

### **Insight:**

- **Experience:** Lead roles have a median salary ~$24,0000 vs ~$15,0000 for entry-level.
   Executive roles show the widest spread, suggesting equity and bonus components vary heavily at the top.
- **Education:** Doctoral and Master's holders have a median salary ~$20,0000 vs ~$18,0000 for Bachelor's,
  though the Bachelor's-to-Master's gap is narrower than often assumed — skills and
  experience are strong equalising factors.
- **Country:** Significant geographic pay disparity exists. US-based roles lead,
  consistent with Silicon Valley compensation norms. Always compare offers after
  adjusting for purchasing-power parity when considering international positions.
- **Company size:** Larger companies pay more on average, reflecting better-funded
  compensation budgets and structured pay bands. Mid-sized firms can be competitive
  when equity upside is factored into total compensation.

```python
# Salary Distribution by work type

plt.figure(figsize=(10,6))

sns.boxplot(x='remote_work', y='annual_salary_usd', data=df)
plt.title("Salary Distribution by Work Type")
plt.ylabel("Average Salary (USD)")
plt.xlabel("Work Arrangement")

plt.tight_layout()
plt.show()
```

### **Insight:**
Fully remote roles are broadly salary-competitive with on-site positions as both have median salary ~$18,0000. This likely reflects that remote AI roles are predominantly offered by larger, higher-paying technology companies with the infrastructure to support distributed teams. Hybrid arrangements sit in a middle range and may offer the best of both worlds for candidates who value flexibility.

```python
# Salary Distribution by LLM Role

plt.figure(figsize=(10,6))
df["llm_label"] = df["is_llm_role"].map({1: "LLM Role", 0: "Non-LLM Role"})

sns.boxplot(x='llm_label', y='annual_salary_usd', data=df)
plt.title("Salary Distribution by LLM Role")
plt.ylabel("Average Salary (USD)")
plt.xlabel("LLM Role")

plt.tight_layout()
plt.show()
```

### **Insights**
LLM-related positions offer a median salary ~$19,0000 vs ~$185,0000for  non-LLM roles, indicating strong market demand for expertise in Generative AI and Large Language Models.

```python
# Demand Score

plt.figure(figsize=(16,12))
# Plot 3
plt.subplot(2,1,1)
ax = (df.groupby('job_title')['demand_score'].mean().sort_values().plot(kind='barh'))
plt.title("Average Demand Score by Job title")
plt.ylabel("Average Demand Score")
plt.xlabel("Job title")
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')

# Plot 4
plt.subplot(2,1,2)
ax = (df.groupby('industry')['demand_score'].mean().sort_values().plot(kind='barh'))
plt.title("Average Demand Score by Industry")
plt.ylabel("Average Demand")
plt.xlabel("Industry")
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')

plt.tight_layout()
plt.show()
```

### **Insight:**
Industries with higher demand scores are investing more aggressively in AI talent. These sectors are likely experiencing rapid AI adoption and expanding their workforce accordingly.

```python
# Top10 Fastest Growing AI Roles

top_roles = (df.groupby('job_title')['demand_growth_yoy_pct']
               .mean()
               .sort_values(ascending=False)
               .head(10))
plt.figure(figsize=(10,6))

ax = (top_roles.sort_values().plot(kind='barh'))

plt.title("Top 10 Fastest Growing AI Roles")
plt.xlabel("Average Demand Growth YoY (%)")
plt.ylabel("Job Title")
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')

plt.tight_layout()
plt.show()
```

### **Insight:**
The fastest-growing AI roles represent emerging areas of opportunity within the industry. Professionals developing skills in these roles are likely to benefit from increasing hiring demand in the coming years.

```python
# Benefit Score

top_roles = (df.groupby('industry')['benefits_score_10']
               .mean()
               .sort_values(ascending=False)
               .head(10))
plt.figure(figsize=(10,6))

ax = (top_roles.sort_values().plot(kind='barh'))

plt.title("Industry Wise Benefits Score")
plt.xlabel("Average Benefits Score")
plt.ylabel("Industry")

for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')

plt.tight_layout()
plt.show()
```

### **Insight:**
The comparison of benefits scores indicates that employee compensation extends beyond salary. Organizations offering stronger benefits packages may be more competitive in attracting and retaining AI talent.

```python
# Job Posting Trend

df['posting_year'] = df['posting_year'].astype(str)
plt.figure(figsize=(12,6))

ax = sns.countplot(
    x='posting_month',
    hue='posting_year',
    data=df,
    order=sorted(df['posting_month'].unique())
)

plt.title("Job Posting Trend (2025 vs 2026)")
plt.xlabel("Posting Month")
plt.ylabel("Number of Job Postings")
plt.legend(title="Year")

for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')
    
plt.tight_layout()
plt.show()
```

### **Insight:**
The job posting trend reveals a significant increase in hiring activity over time. Compared to the same period in 2025, AI job postings during the first three months of 2026 increased by ~85%, indicating a substantial rise in recruitment demand. This upward trend reflects the rapid expansion of AI adoption across industries and highlights the growing need for skilled AI professionals.

```python
# TOP20 in demand skills

skills = (df['required_skills']
          .str.split('|')
          .explode()
          .str.strip().value_counts().head(20))

plt.figure(figsize=(10,6))

ax = skills.sort_values().plot(kind='barh')

plt.title("Most In Demand Skills")
plt.xlabel("Skill Count")
plt.ylabel("Skills")

for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')
    
plt.tight_layout()
plt.show()
```

### **Insight:**
Python, SQL and Cloud dominate skill requirements — they are effectively table-stakes for any AI role. Cloud platforms (AWS, GCP, Azure) appear prominently, confirming that AI deployment is now inseparable from cloud infrastructure. Candidates who combine core ML skills with cloud and MLOps proficiency command the best compensation outcomes and are most resilient to role changes.

# **Conclusion**

### Summary of key findings

| Theme | Finding |
|-------|---------|
| **Salary drivers** | Experience level is the strongest salary driver across all titles. Country of posting is the second-largest driver — US-based roles command the highest absolute salaries. |
| **LLM premium** | LLM-specific roles attract a meaningful salary premium over comparable non-LLM AI roles, consistent across all experience levels. |
| **Remote parity** | Fully remote AI roles are salary-competitive with on-site roles — there is no consistent remote pay penalty in this sector. |
| **Fastest-growing roles** | Roles at the AI application frontier (LLM engineering, AI safety, MLOps) show the strongest year-over-year posting growth. |
| **Must-have skills** | Python, Machine Learning, and cloud platform proficiency appear in the largest share of postings and should be considered baseline requirements. |
| **Benefits** | Technology and Finance industries offer the highest combined salary + benefits packages — most attractive for total compensation. |

---

### Recommendations

**For job seekers:**
- Prioritise Python, cloud (AWS/GCP/Azure), and MLOps skills — these appear in the
  largest share of postings and are positively correlated with salary outcomes.
- Upskill in LLM-specific areas (fine-tuning, RLHF, prompt engineering) to access
  the emerging salary premium documented in this analysis.
- Consider remote-first roles at large technology companies — they offer competitive
  pay without geographic constraints, particularly for candidates outside the US.

**For hiring teams & workforce planners:**
- Demand is highest for senior and specialist roles; allocate longer time-to-fill
  budgets and competitive compensation bands for these positions.
- Benefits quality correlates with offer acceptance; sectors competing on salary alone
  risk losing candidates to those with stronger total compensation packages.
- YoY growth in postings is outpacing the supply of qualified candidates in several
  specialisms. Proactive talent pipelines (internship programmes, upskilling
  partnerships) can reduce dependency on the open market and lower recruitment costs.lopments.
